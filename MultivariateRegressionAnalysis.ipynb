```python
import statsmodels.formula.api as smf
lin_model = smf.ols("Income ~ Limit + Rating + Cards + Age + Education + Balance", data=df).fit()
lin_model.summary()
```

```python
import statsmodels.formula.api as smf
lin_model = smf.ols("Income ~ Limit + Rating + Age +  Balance", data=df).fit()
lin_model.summary()
```

```python
import statsmodels.formula.api as smf
lin_model = smf.ols("Income ~ Limit + Rating +   Balance", data=df).fit()
lin_model.summary()
```

```python
temp_data = pd.DataFrame(dict(fitted_values=lin_model.fittedvalues, residual=lin_model.resid))
graph = sns.lmplot(x='fitted_values', y='residual', data=temp_data, lowess=True, line_kws={'color':'red'})
plt.show()
```

```python
temp_data = pd.DataFrame(dict(Limit=df.Limit, residual=lin_model.resid))
graph = sns.lmplot(x='Limit', y='residual', data=temp_data, lowess=True, line_kws={'color':'red'})
```

```python
temp_data = pd.DataFrame(dict(Rating=df.Rating, residual=lin_model.resid))
graph = sns.lmplot(x='Rating', y='residual', data=temp_data, lowess=True, line_kws={'color':'red'})
```

```python
temp_data = pd.DataFrame(dict(Balance=df.Balance, residual=lin_model.resid))
graph = sns.lmplot(x='Balance', y='residual', data=temp_data, lowess=True, line_kws={'color':'red'})
```

```python
sns.distplot(lin_model.get_influence().resid_studentized_internal)
plt.show()
```

```python
import statsmodels.formula.api as smf
lin_model = smf.ols("Income ~ Limit + Rating +   Balance + np.square(Balance)", data=df).fit()
lin_model.summary()
```

```python
temp_data = pd.DataFrame(dict(fitted_values=lin_model.fittedvalues, residual=lin_model.resid))
graph = sns.lmplot(x='fitted_values', y='residual', data=temp_data, lowess=True, line_kws={'color':'red'})
plt.show()
```

```python
temp_data = pd.DataFrame(dict(Limit=df.Limit, residual=lin_model.resid))
graph = sns.lmplot(x='Limit', y='residual', data=temp_data, lowess=True, line_kws={'color':'red'})
```

```python
temp_data = pd.DataFrame(dict(Rating=df.Rating, residual=lin_model.resid))
graph = sns.lmplot(x='Rating', y='residual', data=temp_data, lowess=True, line_kws={'color':'red'})
```

```python
import statsmodels.formula.api as smf
lin_model = smf.ols("Income ~ Limit + np.square(Limit)+  Balance ", data=df).fit()
lin_model.summary()
```

```python
temp_data = pd.DataFrame(dict(fitted_values=lin_model.fittedvalues, residual=lin_model.resid))
graph = sns.lmplot(x='fitted_values', y='residual', data=temp_data, lowess=True, line_kws={'color':'red'})
plt.show()
```

```python
import statsmodels.formula.api as smf
lin_model = smf.ols("Income ~ Limit + np.square(Limit)+ np.power(Limit,3)+ Balance ", data=df).fit()
lin_model.summary()
```

```python
import statsmodels.formula.api as smf
lin_model = smf.ols("Income ~  np.square(Limit)+ np.power(Limit,3)+ Balance ", data=df).fit()
lin_model.summary()
```

```python
temp_data = pd.DataFrame(dict(fitted_values=lin_model.fittedvalues, residual=lin_model.resid))
graph = sns.lmplot(x='fitted_values', y='residual', data=temp_data, lowess=True, line_kws={'color':'red'})
plt.show()
```

```python
import statsmodels.formula.api as smf
lin_model = smf.ols("Income ~  np.power(Limit,1/2)+ np.power(Limit,3)+ Balance ", data=df).fit()
lin_model.summary()
```

```python
import statsmodels.formula.api as smf
lin_model = smf.ols("Income ~  Student+np.square(Limit)+  Balance ", data=df).fit()
lin_model.summary()
```

```python
temp_data = pd.DataFrame(dict(fitted_values=lin_model.fittedvalues, residual=lin_model.resid))
graph = sns.lmplot(x='fitted_values', y='residual', data=temp_data, lowess=True, line_kws={'color':'red'})
plt.show()
```
